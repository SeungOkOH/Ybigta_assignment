1번 시각화 선택

```python
# 여기에 공통과제 2번 코드를 작성해주세요. 주석을 꼼꼼하게 작성하여 왜 이렇게 시각화를 했는지 설명해주시면 더 좋습니다!
zipped_data = zip(df['carat'], df['cut'], df['color'], df['clarity'], df['price'])
cmap = plt.colormaps["Purples"]

fig, ax = plt.subplots()
fig.suptitle("Carat vs. Price for Color 'D'", size=14)

# D 색상에 대해 진행. 필요시 다른 색상으로 변경하면 됨.
selected_color = 'D'
color_scatter = {'x': [], 'y': []}

for carat, cut, color, clarity, price in zipped_data:
    if color == selected_color:
        color_scatter['x'].append(carat)
        color_scatter['y'].append(price)

ax.scatter(color_scatter['x'], color_scatter['y'], label=selected_color, color=cmap(0.5), alpha=0.9, s=5)
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=None, cmap=cmap), ax=ax, label="carat")
ax.set_xlabel("Carat")
ax.set_ylabel("Price")
ax.set_ylim(0, 5000)
ax.legend()

plt.show()
```

```python
# 어떤 요인이 마진을 남기는데 요인이 되는지 알고 싶음
# Y = (Sale Price - Retail Price), X(후보) = Brand, Sneaker Name, Shoe Size, Buyer Region

# 예시: 발 사이즈, 브랜드와 마진의 관계
df['Sale Price'] = df['Sale Price'].replace('[\$,]', '', regex=True).astype(float)
df['Retail Price'] = df['Retail Price'].replace('[\$,]', '', regex=True).astype(float)
df['margin'] = df['Sale Price'] - df['Retail Price']

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Shoe Size', y='margin', hue='Brand', data=df, alpha=0.7, s=50)
plt.xlabel('Shoe Size')
plt.ylabel('Margin')
plt.title('Scatter Plot of Shoe Size vs. Margin by Brand')
plt.legend(title='Brand')
plt.show()
```
