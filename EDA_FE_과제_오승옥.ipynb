### 5) Feature Engineering

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12
missing_percent = df.isnull().mean()

# 결측치가 25% 이상인 column들 제거
drop_cols_25_percent = missing_percent[missing_percent >= 0.25].index
df = df.drop(columns=drop_cols_25_percent)

# 문자열이 포함된 column 제거
string_cols = df.select_dtypes(include='object').columns
df = df.drop(columns=string_cols)

# 결측치가 10% 이상 25% 미만인 row들 제거
drop_rows_10_25_percent = df[df.isnull().sum(axis=1) >= 0.10 * len(df)].index
df = df.drop(index=drop_rows_10_25_percent)

# 열 삭제 후 다시 결측치 비율 확인
missing_percent = df.isnull().mean()

# 결측치가 10% 미만인 column들에 대해 중앙값으로 대체
fill_cols_10_percent = missing_percent[(missing_percent < 0.10) & (missing_percent > 0)].index
df[fill_cols_10_percent] = df[fill_cols_10_percent].apply(lambda x: x.fillna(x.median()))

# 결과 확인
df.info()
```
