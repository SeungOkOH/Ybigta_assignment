```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import requests
print(selenium.__version__)
```

```python
columns = ['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers']
codimap_df = pd.DataFrame(columns=columns)
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1'
res = driver.get(url)
driver.implicitly_wait(2) 
try:
    driver.maximize_window()
except:
    pass
```

```python
interval = 1
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
        # 스크롤을 가장 아래로 내림
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

        # 페이지 로딩 대기
        time.sleep(interval)

        # 현재 문서 높이를 가져와서 저장
        curr_height = driver.execute_script("return document.body.scrollHeight")
        if curr_height == prev_height:
            break

        prev_height = curr_height
```

```python
codimap_category = []
codimap_title = []
codimap_date = []
views = []
comment_numbers = []

soup = BeautifulSoup(driver.page_source, 'html.parser')

span_tags = soup.find_all('span', {'class': 'style-list-information__text'})
codimap_category = [span.text for span in span_tags[:60]]

strong_tags = soup.find_all('strong', {'class': 'style-list-information__title'})
codimap_title = [strong.text for strong in strong_tags[:60]]

date_tags = soup.find_all('span', {'class': 'post-information__text'})
date_view_comment = [date.text for date in date_tags[:180]]

for i in range(60):
    # 날짜
    codimap_date.append(date_view_comment[i * 3])

    # 조회수
    views_index = i * 3 + 1
    views.append(date_view_comment[views_index])

    # 댓글 수
    comment_numbers_index = i * 3 + 2
    comment_numbers.append(date_view_comment[comment_numbers_index])

views = [view[3:] for view in views]
comment_numbers = [comment_number[3:] for comment_number in comment_numbers]
   
print(codimap_category)
print(codimap_title)
print(codimap_date)
print(views)
print(comment_numbers)
```

```python
codimap_df = pd.DataFrame({
    'codimap_category': codimap_category,
    'codimap_title': codimap_title,
    'codimap_date': codimap_date,
    'views': views,
    'comment_numbers': comment_numbers
})
```
